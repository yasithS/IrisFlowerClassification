### Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pandas as pd
%matplotlib inline
```

### Loading the dataset

```python
columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Class_labels']
# load the data
df = pd.read_csv('Data/iris.data', names=columns)
df.head()
```

### Visualization

```python
df.describe()
```

```python
# visualize the whole dataset
sns.pairplot(df, hue='Class_labels')
```

### Separating input columns and output column

```python
# separate features and target
data = df.values

X = data[:, 0:4]
Y = data[:,4]

# check
print(X)
```

### Splitting the data into training and testing

```python
### splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

### support vector machine algorithm

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, y_train)
```

```python
svc_prediction = model_svc.predict(X_test)
# calculate the accuaracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, svc_prediction)*100)
```

```python
# checking the actual values and prediction values
for i in range(len(svc_prediction)):
  print(y_test[i], svc_prediction[i])
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
logistic_prediction = model_lr.predict(X_test)
# calculate the accuaracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, logistic_prediction)*100)
```

```python
# checking the actual values and prediction values
for i in range(len(logistic_prediction)):
  print(y_test[i], logistic_prediction[i])
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train, y_train)
```

```python
dt_prediction = model_dt.predict(X_test)
# calculate the acuaracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, dt_prediction)*100)
```

```python
# checking the actual values and prediction values
for i in range(len(dt_prediction)):
  print(y_test[i], dt_prediction[i])
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, svc_prediction))
```
